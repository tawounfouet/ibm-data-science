### The first one will be a plot to summarize the results of a survey that was conducted to gauge an audience interest in different data science topics. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('topic.csv', index_col = 'Unnamed: 0')
```

```python
df
```

```python
df1 = df.sort_values(ascending=False, by = 'Very interested')
```

```python
df1.head()
```

```python
df1['Very interested'] = df1['Very interested'].div(2233)
```

```python
df1['Somewhat interested'] = df1['Somewhat interested'].div(2233)
```

```python
df1['Not interested'] = df1['Not interested'].div(2233)
```

```python
df1
```

```python
df1 = df1 * 100
```

```python
df1
```

```python
df1.round(2)
df1
```

```python
df2 = df1.round({'Very interested': 2, 'Somewhat interested': 2, 'Not interested': 2})
```

```python
df2  #Finalised dataframe
```

```python
df2.reset_index(inplace=True)
df2
```

```python
df2.rename(columns={"index": "Topics"}, inplace=True)
```

```python
df2
```

```python
ax = df2.plot.bar(x='Topics', figsize=(20,8), width=0.8, fontsize =14, color=['#5cb85c','#5bc0de','#d9534f'])

for p in ax.patches:
    ax.annotate(str(p.get_height()), xy=(p.get_x(), p.get_height()))



plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)
plt.legend(fontsize =14)
plt.xlabel('')
plt.yticks([])
plt.show()
```
